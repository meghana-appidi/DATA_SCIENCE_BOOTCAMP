## 1. Create a line plot of ZN and INDUS in the housing data.

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.
	b. Change the figure size to a width of 12 and height of 8.
	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Boston housing data
housing_csv = 'YOUR-REPOSITORY/HousingData.csv'
housing = pd.read_csv(housing_csv)

# Change the style sheet
plt.style.use('fivethirtyeight')

# Create a line plot of ZN and INDUS
plt.figure(figsize=(12, 8))
plt.plot(housing['ZN'], color='green', linestyle='-', label='ZN')  # Solid green line for ZN
plt.plot(housing['INDUS'], color='blue', linestyle='--', label='INDUS')  # Blue dashed line for INDUS
plt.title("Line Plot of ZN and INDUS in Boston Housing Data")
plt.legend()
plt.show()
```

## 2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.
	b. Move the legend to the lower-left corner.
	c. Do the same thing but with horizontal bars.
	d. Move the legend to the upper-right corner.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating dummy data
dummy_data = pd.DataFrame({
    'col1': [5, 10, 15, 20, 25],
    'col2': [30, 25, 20, 15, 10]
})

# Vertical bar chart
plt.figure(figsize=(10, 6))
dummy_data.plot(kind='bar')
plt.title("Dummy Data Vertical Bar Chart")  # Large title
plt.legend(loc='lower left')  # Legend in the lower-left corner
plt.show()

# Horizontal bar chart
plt.figure(figsize=(10, 6))
dummy_data.plot(kind='barh')
plt.title("Dummy Data Horizontal Bar Chart")  # Large title
plt.legend(loc='upper right')  # Legend in the upper-right corner
plt.show()
```

## 3. Create a histogram with pandas for using MEDV in the housing data.

	a. Set the bins to 20


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Boston housing data
housing_csv = 'YOUR-REPOSITORY/HousingData.csv'  # Replace with the path to your CSV file
housing = pd.read_csv(housing_csv)

# Create a histogram for MEDV with 20 bins
plt.figure(figsize=(10, 6))
housing['MEDV'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of MEDV in Boston Housing Data")
plt.xlabel("MEDV")
plt.ylabel("Frequency")
plt.show()
```

## 4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Boston housing data
housing_csv = 'YOUR-REPOSITORY/HousingData.csv' 
housing = pd.read_csv(housing_csv)

# Create a scatter plot of two features with a very positive correlation


# Generating a heatmap for the correlation matrix
plt.figure(figsize=(12, 10))
corr_matrix = housing.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Boston Housing Data")
plt.show()

# Based on the heatmap, select two features with strong positive correlation
# For example, if RM and MEDV show strong positive correlation

plt.figure(figsize=(10, 6))
sns.scatterplot(x=housing['RM'], y=housing['MEDV'])
plt.title("Scatter Plot of RM and MEDV")
plt.xlabel("RM (Average Number of Rooms per Dwelling)")
plt.ylabel("MEDV (Median Value of Owner-Occupied Homes)")
plt.show()
```
